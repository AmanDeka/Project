```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/drug200.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
x=df.iloc[:,0:5]
x.head()
```

```python
y=pd.get_dummies(df.iloc[:,5:]).values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
x.shape, y.shape
```

```python
xtrain.shape,xtest.shape
```

```python
df['BP'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
xtrain['Sex'] = le.fit_transform(xtrain['Sex'])
xtrain['Sex']
xtest['Sex'] = le.transform(xtest['Sex'])
xtest['Sex']
xtrain['BP'] = le.fit_transform(xtrain['BP'])
xtrain['BP']
xtest['BP'] = le.transform(xtest['BP'])
xtest['BP']
xtrain['Cholesterol'] = le.fit_transform(xtrain['Cholesterol'])
xtrain['Cholesterol']
xtest['Cholesterol'] = le.transform(xtest['Cholesterol'])
xtest['Cholesterol']
```

```python
xtrain
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(5,input_dim=5,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(5,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=19,batch_size=5,validation_data=(xtest,ytest))
```

```python
v=np.random.rand(10,5)
v
```

```python
z=model.predict(v)
z=z.round(decimals=0)
z=z.astype(int)
z
```

```python
class_labels = np.argmax(z, axis=1)
class_labels
```
